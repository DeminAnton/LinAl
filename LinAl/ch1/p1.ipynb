### Вектор

Поскольку существует множество способов представления векторов, изначально Стрэнг использует вектор-строки и вектор-столбцы. Вообще, векторы необязательно представлять в виде наборов координат - они могут существовать и вне координатной системы. Важно понимать, что вектор - это объект, который можно определить независимо от координатной системы. Он описывается своими свойствами и сохраняет свои инварианты, такие как длина, а представление в виде набора координат — это лишь один из способов его записи.

$$
\boldsymbol{v} =
\begin{bmatrix}
v_1 \\
v_2
\end{bmatrix}
\quad
\begin{aligned}
v_1 &= \text{первый компонент } \boldsymbol{v} \\
v_2 &= \text{второй компонент } \boldsymbol{v}
\end{aligned}
$$

И тут можно сразу ввести те самые операции, которые делают вектор вектором:

1. **Сложение двух векторов.** Это такая операция, в результате которой получается ещё один вектор.  
   Для вектор-столбцов эта операция выглядит так:
   $$
   \boldsymbol{v} + \boldsymbol{w} = 
   \begin{bmatrix} 
   v_1 \\
   v_2
   \end{bmatrix}
   +
   \begin{bmatrix} 
   w_1 \\
   w_2
   \end{bmatrix}
   =
   \begin{bmatrix} 
   v_1 + w_1\\
   v_2 + w_2
   \end{bmatrix}
   $$

2. **Умножение вектора на скаляр** (скаляр — это одномерный вектор или, попросту, число):
   $$
   c \boldsymbol{v} = 
   c
   \begin{bmatrix} 
   v_1 \\
   v_2
   \end{bmatrix}
   =
   \begin{bmatrix} 
   c v_1 \\
   c v_2
   \end{bmatrix}
   $$
